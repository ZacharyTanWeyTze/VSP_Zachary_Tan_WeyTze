## 5. Descriptive Statistics and Data Visualization (Global)

Data Description For Global Indicators/Variables
* Adult literacy rate is the percentage of people ages 15 and above who can both read and write with understanding a short simple statement about their everyday life.
* GDP per capita is gross domestic product divided by midyear population. GDP is the sum of gross value added by all resident producers in the economy plus any product taxes and minus any subsidies not included in the value of the products. It is calculated without making deductions for depreciation of fabricated assets or for depletion and degradation of natural resources. Data are in current U.S. dollars.
* Gini index measures the extent to which the distribution of income (or, in some cases, consumption expenditure) among individuals or households within an economy deviates from a perfectly equal distribution. A Gini index of 0 represents perfect equality, while an index of 100 implies perfect inequality.
* General government expenditure on education (current, capital, and transfers) is expressed as a percentage of total general government expenditure on all sectors (including health, education, social services, etc.). It includes expenditure funded by transfers from international sources to government.
* Primary school pupil-teacher ratio is the average number of pupils per teacher in primary school.
* Net enrollment rate is the ratio of children of official school age who are enrolled in school to the population of the corresponding official school age. Primary education provides children with basic reading, writing, and mathematics skills along with an elementary understanding of such subjects as history, geography, natural science, social science, art, and music.

## 6. Regression Modelling (Global)

```python
# Import linear model and mse
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

## 6.1 Linear Model 1 - Simple LR with One X Variable

```python
# Define x and y for LR Model 1
X1 = Global[['Pupil Teacher Ratio Primary (10Y AVG)']]
y = Global[['Literacy Rate Adult % (10Y AVG)']]

# Intialize LR Model 1 (Simple LR with 1 "x" variable)
lr_model_1 = linear_model.LinearRegression()
lr_model_1.fit(X1, y)
```

```python
# Calculate the coefficients and metrics
constant = lr_model_1.intercept_
X1_coefficient = lr_model_1.coef_[0]

# Calculate metrics
r_squared = lr_model_1.score(X1, y)
mse = mean_squared_error(y, lr_model_1.predict(X1))

# Print the LR Model 1 equation, r-squared value, and MSE
print(f"LR Model 1: y = {constant[0]:.4f} {X1_coefficient[0]:.4f}(X1)")
print("where y = Literacy Rate Adult %")
print("where X1 = Pupil Teacher Ratio Primary")
print(f"R-Squared: {r_squared:.4f}")
print(f"Mean Squared Error: {mse:.4f}")
```

```python
# Scatter Plot with LR Model 1
# Create scatter plot
plt.scatter(X1, y, color = 'grey', marker = 'd')

# Add regression line
plt.plot(X1, lr_model_1.predict(X1), color = 'black')

# Add title and labels
plt.xlabel('Pupil Teacher Ratio, Primary')
plt.ylabel('Literacy Rate Adult %')
plt.title('Literacy Rate vs Pupil Teacher Ratio')
plt.show()
```

LR Model 1 Interpretation:
* Intercept (115.0382): This is the predicted literacy rate when the pupil-teacher ratio is zero. While a pupil-teacher ratio of zero is not realistic, the intercept provides a starting point for the regression line.
* Slope (-1.2273): For every one unit increase in the pupil-teacher ratio, the literacy rate is expected to decrease by 1.2273 percentage points. Higher pupil-teacher ratios are associated with lower literacy rates.


## 6.2 Linear Model 2 - Multiple Linear Regression With Four X Variables

```python
# Define x and y for LR Model 2
multi_X = Global.drop(columns=['Literacy Rate Adult % (10Y AVG)']) # drop target variable from global data
multi_X = multi_X.drop(columns=['Country Name']) # drop country name from global data, taking remaining columns as variables
y = Global['Literacy Rate Adult % (10Y AVG)']

# Intialize LR Model 2 (Mutliple LR with 4 "x" variables)
lr_model_2 = linear_model.LinearRegression()
lr_model_2.fit(multi_X, y)
```

```python
# Calculate the coefficients and metrics
constant = lr_model_2.intercept_
X1_coefficient = lr_model_2.coef_[0]
X2_coefficient = lr_model_2.coef_[1]
X3_coefficient = lr_model_2.coef_[2]
X4_coefficient = lr_model_2.coef_[3]

# Calculate metrics
r_squared = lr_model_2.score(multi_X, y)
mse = mean_squared_error(y, lr_model_2.predict(multi_X))

# Print the LR Model 2 equation, r-squared value, and MSE
print(f"LR Model 2: y = {constant:.4f} {X1_coefficient:.4f}(X1) {X2_coefficient:.4f}(X2) + {X3_coefficient:.4f}(X3) {X4_coefficient:.4f}(X4)")
print("where y = Literacy Rate Adult %")
print("where X1 = Pupil Teacher Ratio Primary")
print("where X2 = Government Expenditure on Education as % of Total Expenditure")
print("where X3 = School Enrollment Primary % Net")
print("where X4 = GDP Per Capita Current USD")
print(f"R-Squared: {r_squared:.4f}")
print(f"Mean Squared Error: {mse:.4f}")
```

```python
# Check LR Model 2 coefficients
coefficients = lr_model_2.coef_
feature_names = multi_X.columns
for i in range(len(coefficients)):
  print(f"{feature_names[i]}: {coefficients[i]}")
```

LR Model 2 Interpretation:
* Intercept (46.5460): This is the predicted literacy rate when all predictor variables are zero. Again, while these values may not be realistic, the intercept provides a starting point for the regression line.
* Slope (-0.9017 for (X1)): For every one unit increase in the pupil-teacher ratio, the literacy rate is expected to decrease by 0.9017 percentage points.
* Slope (-0.3080 for (X2)): For every one unit increase in government expenditure on education as a percentage of total expenditure, the literacy rate is expected to decrease by 0.3080 percentage points.
* Slope (0.7244 for (X3)): For every one unit increase in school enrollment percentage, the literacy rate is expected to increase by 0.7244 percentage points.
* Slope (-0.0000 for (X4)): The coefficient for GDP per capita is very close to zero, indicating it has a negligible effect on literacy rate in this model.



## Summary/Comparison of Regression Analysis
LR Model 2 outperforms LR Model 1 in both R-Square and MSE.
* LR Model 2 explains more variability in the literacy rate.
* LR Model 2 R-Squared value means that 72.69% of the variability in the literacy rate can be explained by the four predictor variables.
* LR Model 1 R-Squared value means that 60.40% of the variability in the literacy rate can be explained by the pupil-teacher ratio (primary).
* LR Model 2 has better fit and predictive accuracy (83 MSE VS 120 MSE).
